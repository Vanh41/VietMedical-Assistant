```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("NgVietAnh41/Qwen3-4b-it-VietMedQA")
model = AutoModelForCausalLM.from_pretrained("NgVietAnh41/Qwen3-4b-it-VietMedQA")
```

```python


from llama_cpp import Llama

llm = Llama(
    model_path="model.gguf",
    verbose=False
)


```

```python
# stream chat completion using existing llm and defaut_generator_config
for chunk in llm.create_chat_completion(
	messages=[{"role": "user", "content": "T đau đầu quá"}],
	stream=True,
):
	choices = chunk.get("choices", [])
	if not choices:
		continue
	choice = choices[0]
	# streaming content may appear under delta.content or text depending on version
	content = choice.get("delta", {}).get("content") or choice.get("text") or chunk.get("text")
	if content:
		print(content, end="", flush=True)
	if choice.get("finish_reason"):
		break
print()
```
